```python
import pandas as pd
import matplotlib.pyplot as plt
from faker import Faker

def generate_medical_data(num_patients):
    fake = Faker()
    data = {'Name': [], 'Age': [], 'Gender': [], 'Condition': [], 'Treatment': [], 'Outcome': []}
    for _ in range(num_patients):
        data['Name'].append(fake.name())
        data['Age'].append(fake.random_int(min=18, max=100, step=1))
        data['Gender'].append(fake.random_element(elements=('Male', 'Female')))
        data['Condition'].append(fake.random_element(elements=('Asthma', 'Diabetes', 'Hypertension', 'COVID-19', 'Cancer')))
        data['Treatment'].append(fake.random_element(elements=('Medication', 'Surgery', 'Physiotherapy', 'Chemotherapy', 'Radiotherapy')))
        data['Outcome'].append(fake.random_element(elements=('Recovered', 'Improved', 'Stable', 'Worsened', 'Died')))
    return pd.DataFrame(data)

medical_data = generate_medical_data(1000)
medical_data.to_csv('medical_data.csv', index=False)

medical_data = pd.read_csv('medical_data.csv')

condition_counts = medical_data['Condition'].value_counts()
condition_counts.plot(kind='bar')
plt.title('Number of Patients by Condition')
plt.xlabel('Condition')
plt.ylabel('Number of Patients')
plt.show()

condition_age = medical_data.groupby('Condition')['Age'].mean()
condition_age.plot(kind='bar')
plt.title('Average Age by Condition')
plt.xlabel('Condition')
plt.ylabel('Average Age')
plt.show()

outcome_counts = medical_data['Outcome'].value_counts()
outcome_counts.plot(kind='bar')
plt.title('Number of Patients by Outcome')
plt.xlabel('Outcome')
plt.ylabel('Number of Patients')
plt.show()

gender_counts = medical_data['Gender'].value_counts()
gender_counts.plot(kind='bar')
plt.title('Number of Patients by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Patients')
plt.show()
```
